```python
# concatenate all the dataframes but filter for DC
df = pd.concat([_2001_2002, _2003_2004, _2005_2007, _2008_2009, _2010_2011, _2012, _2013, _2014, _2015, _2016, _2017, _2018, _2019, _2020, _2021, _2022])
```

```python
df.columns
```

```python
# filter for DC
df = df[df['STATE'] == 'DC']
# Fill NaN values in column1 with values from column2
df['GENERATION (Megawatthours)'].fillna(df['GENERATION\n(Megawatthours)'], inplace=True)
# drop column2
df.drop(['GENERATION\n(Megawatthours)'], axis=1, inplace=True)
```

```python
df
```

```python
# delete all the rows with Total in "energy source" column
df = df[df['ENERGY SOURCE'] != 'Total']
```

```python
# delete total electric power industry rows
df = df[df['TYPE OF PRODUCER'] != 'Total Electric Power Industry']
```

```python
df[-12:-1]
```

```python
# Assuming your DataFrame is named 'df'
# Pivot the DataFrame to have 'Type of Producer' as columns
type_of_producer_df = df.pivot_table(index=['YEAR', 'MONTH'], columns='TYPE OF PRODUCER', values='GENERATION (Megawatthours)', aggfunc=lambda x: 0 if pd.isnull(x).all() else x).fillna(0)

# Pivot the DataFrame to have 'Energy Source' as columns
energy_source_df = df.pivot_table(index=['YEAR', 'MONTH'], columns='ENERGY SOURCE', values='GENERATION (Megawatthours)', aggfunc=lambda x: 0 if pd.isnull(x).all() else 1).fillna(0)

# Merge the two DataFrames on the index
result_df = pd.merge(type_of_producer_df, energy_source_df, left_index=True, right_index=True)

# Reset index to make 'YEAR' and 'MONTH' regular columns
result_df.reset_index(inplace=True)

# Display the result
print(result_df)
```

```python
# Assuming your DataFrame is named 'df'
# Pivot the DataFrame to have 'Type of Producer' as columns
type_of_producer_df = df.pivot_table(index=['YEAR', 'MONTH'], columns='TYPE OF PRODUCER', values='GENERATION (Megawatthours)', aggfunc=lambda x: 0 if pd.isnull(x).all() else sum(x) if isinstance(x, list) else x).fillna(0)

# Pivot the DataFrame to have 'Energy Source' as columns
energy_source_df = df.pivot_table(index=['YEAR', 'MONTH'], columns='ENERGY SOURCE', values='GENERATION (Megawatthours)', aggfunc=lambda x: 0 if pd.isnull(x).all() else sum(x) if isinstance(x, list) else x).fillna(0)

# Merge the two DataFrames on the index
result_df = pd.merge(type_of_producer_df, energy_source_df, left_index=True, right_index=True)

# Reset index to make 'YEAR' and 'MONTH' regular columns
result_df.reset_index(inplace=True)

# Display the result
print(result_df)
```

```python
# Function to sum arrays or return the original value
def sum_arrays_or_value(x):
    if isinstance(x, (list, np.ndarray)):
        return np.sum(x)
    return x

# Apply the function to each column in the DataFrame
for col in df.columns:
    df[col] = df[col].apply(sum_arrays_or_value)

# Assuming the previous code for merging and resetting index
result_df = pd.merge(type_of_producer_df, energy_source_df, left_index=True, right_index=True)
result_df.reset_index(inplace=True)
```
